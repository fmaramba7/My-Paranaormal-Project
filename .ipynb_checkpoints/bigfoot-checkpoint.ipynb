```python
# import dependencies
# !pip install plotly.plotly
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
import plotly as py
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
# Read dataset
df = pd.read_csv('Resources/bfro-report-locations.csv')
df.head()
```

```python
# split the timestamp to date and time columns
df['dates'] = pd.to_datetime(df['timestamp']).dt.date
df['time'] = pd.to_datetime(df['timestamp']).dt.time
df.head()
```

```python
#split title column into two columns
df[['Report', 'description']] = df['title'].str.split(':', 1, expand=True)
df
```

```python
# copy columns needed for project
df_clean = df[['number','dates','time','latitude','longitude','description']].copy()
df_clean
```

```python
# save clean df to Resources folder
df_clean.to_csv('Resources/clean_data.csv', index=False)
```
